```python
import os
import pandas as pd
import numpy as np
```

```python
csv_path = os.path.expanduser('~/Downloads/permits.csv')
df = pd.read_csv(csv_path)
print(f'Loaded {csv_path} with {len(df)} rows')
```

```python
df = df.dropna(subset=['Filing Date'])
df['filing_date_datetime'] = pd.to_datetime(df['Filing Date'], errors='coerce')

df.sort_values('filing_date_datetime', inplace=True)
```

```python
splits = np.array_split(df, 5)
out_dir = "../data/partitioned"

for i, part in enumerate(splits, 1):
    out_path = os.path.join(out_dir, f'permits_part_{i}.csv')
    part.to_csv(out_path, index=False)
    print(f'Wrote {out_path} with {len(part)} rows')
```

```python
import shutil
from pathlib import Path

data_dir = Path("../data/partitioned")
archieve_dir = Path("../data/compressed")

for entry in sorted(data_dir.iterdir()):
    archive_base = archieve_dir / entry.name  # archive will be data/<folder>.zip
    print(f"Compressing {entry} -> {archive_base.with_suffix('.zip')}")
    shutil.make_archive(str(archive_base), 'zip', root_dir=str(data_dir), base_dir=entry.name)
```

```python
# Unzip files from compressed directory to `unzipped`

unzipped_dir = Path("../data/unzipped")
for entry in sorted(archieve_dir.iterdir()):
    print(f"Unzipping {entry} -> {unzipped_dir}")
    shutil.unpack_archive(str(entry), extract_dir=str(unzipped_dir)) 
```

```python
# Read in and concatenate all unzipped CSV files
all_dfs = []
for entry in sorted(unzipped_dir.iterdir()):
    if entry.suffix == '.csv':
        print(f"Reading {entry}")
        df_part = pd.read_csv(entry)
        all_dfs.append(df_part)
        full_df = pd.concat(all_dfs, ignore_index=True)

print(f"Concatenated DataFrame has {len(full_df)} rows")
```
